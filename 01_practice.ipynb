```python
def calc(a, operator, b):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        return a / b
    else:
        raise ValueError("유효하지 않은 연산자입니다.")

print(calc(10, '+', 5))  # 15
```

```python
def calcGrade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

score = int(input("성적을 입력하세요: "))
print("학점:", calcGrade(score))
```

```python
def divide(a, b):
    if b == 0:
        raise ValueError("0으로 나눌 수 없습니다.")
    return (a / b, a % b)

t = divide(10, 3)
print(t)
```

```python
# 클래스 정의 실습

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def double(self):
        self.width *= 2
        self.height *= 2

rect = Rectangle(4, 5)
print("넓이:", rect.area())  # 넓이: 20
rect.double()
print(" doubled 넓이:", rect.area())  # doubled 넓이: 80
```

```python
# 클래스 상속 실습
class Vehicle:
    def __init__(self, speed):
        self.speed = speed
    
    def speed_up(self):
        self.speed += 10
    
    def speed_down(self):
        self.speed -= 10

class Car(Vehicle):
    def __init__(self, speed, wheels, seats):
        Vehicle.__init__(self, speed)
        self.wheels = wheels
        self.seats = seats
    
    def info(self):
        print(f"속도: {self.speed}km/h, 바퀴 수: {self.wheels}, 좌석 수: {self.seats}")

class Truck(Car):
    def __init__(self, speed, wheels, seats, loadage):
        Car.__init__(self, speed, wheels, seats)
        self.loadage = loadage
    
    def load(self):
        print("load")

    def unload(self):
        print("unload")
    
    def info(self):
        super().info()
        print(f"적재량: {self.loadage}톤")

t = Truck(80, 6, 2, 10)
t.info()  # 속도: 80km/h, 바퀴 수: 6, 좌석 수: 2, 적재량: 10톤
t.load()   
t.unload() 
```

```python
# datetime module을 사용해서 올해의 남은 일 수와 지금부터 
# 100일 후의 날짜와 시간을 출력하시오.

import datetime
# 현재 날짜와 시간
now = datetime.datetime.now()

# 올해의 마지막 날
end_of_year = datetime.datetime(now.year, 12, 31)

# 올해의 남은 일 수
remaining_days = (end_of_year - now).days

# 100일 후의 날짜와 시간
hundred_days_later = now + datetime.timedelta(days=100)

print(f"올해의 남은 일 수: {remaining_days}")
print(f"100일 후의 날짜와 시간: {hundred_days_later}")
```

```python
# time module과 for문을 사용해서 
# 1~1000까지의 합을 구하는데 경과시간을 출력하시오.

import time
start_time = time.time()
total_sum = 0
for i in range(1, 1001):
    total_sum += i
end_time = time.time()
elapsed_time = end_time - start_time
print(f"1~1000까지의 합: {total_sum}")
print(f"경과 시간: {elapsed_time:.10f}초")
```

```python
# random 모듈을 이용해서 두 가지 방법으로 lotto 발생 함수를 만드시오.

import random

# 1. while 루프와 randint 함수를 이용
# radint(a, b) 함수는 a와 b 사이의 임의의 정수를 반환합니다.
def generate_lotto():
    lotto_numbers = set()
    while len(lotto_numbers) < 6:
        number = random.randint(1, 45)
        lotto_numbers.add(number)
    return sorted(lotto_numbers)


# 2. shuffle 함수를 이용\
# 리스트의 요소들을 무작위로 섞는 함수, 원본 리스트를 직접 변경합니다.
def generate_lotto_shuffle():
    numbers = list(range(1, 46))
    random.shuffle(numbers)
    return sorted(numbers[:6])

# 3. sample 함수를 이용
# sample(population, k) 함수는 모집단에서 k개의 고유한 요소를 선택하여 반환합니다.
def generate_lotto_sample():
    return sorted(random.sample(range(1, 46), 6))

print("Lotto numbers (while + randint):", generate_lotto())
print("Lotto numbers (shuffle):", generate_lotto_shuffle())
print("Lotto numbers (sample):", generate_lotto_sample())
```

```python
# map과 lambda를 이용해서 1~10 list의 각 아이템에 
# 5배를 가진 list를 만들고 출력하시오.

numbers = list(range(1, 11))
result = list(map(lambda x: x * 5, numbers))
print(result)
```

```python
# 1부터 10 사이의 정수를 가진 리스트에서 짝수만 나오게 
# 필터링한 결과 리스트의 모든 원소에 대해 제곱을 수행해서 
# 리스트로 반환하는 코드를 필터와 맵과 람다를 이용해서 작성하시요.

numbers = list(range(1, 11))
even_numbers = filter(lambda x: x % 2 == 0, numbers)
squared_even_numbers = map(lambda x: x ** 2, even_numbers)
result = list(squared_even_numbers)
print(result)
```

```python
# 1부터 10 사이의 정수를 가진 리스트에서 
# 짝수만 나오게 필터링한 결과 리스트의 
# 모든 원소의 곱을 구하는 코드를 
# lambda, filter, reduce 함수를 사용해서 작성하시요.

# reduce 함수는 functools 모듈에서 제공하는 함수로, 
# 리스트의 모든 요소에 대해 누적적으로 함수를 적용하여 
# 단일 값을 반환합니다. 
# ex) reduce(lambda x, y: x + y, [1, 2, 3, 4])는 10을 반환합니다.

from functools import reduce
numbers = list(range(1, 11))
even_numbers = filter(lambda x: x % 2 == 0, numbers)
product_of_even_numbers = reduce(lambda x, y: x * y, even_numbers)
print(product_of_even_numbers)
```

```python
# 축약표현으로 다음의 리스트를 만드시오.

# 1. 1~10까지 숫자의 제곱을 포함하는 리스트
squares = [x**2 for x in range(1, 11)]

# 2. 1~20까지의 짝수를 포함하는 리스트
evens = [x for x in range(1, 21) if x % 2 == 0]

# 3. 1~20까지의 숫자중 3의 배수를 포함하는 리스트
multiples_of_3 = [x for x in range(1, 21) if x % 3 == 0]

# 4. 1~20까지의 숫자중 5의 배수의 제곱을 포함하는 리스트
squares_of_multiples_of_5 = [x**2 for x in range(1, 21) if x % 5 == 0]

print("1~10까지 숫자의 제곱:", squares)
print("1~20까지의 짝수:", evens)
print("1~20까지의 숫자중 3의 배수:", multiples_of_3)
print("1~20까지의 숫자중 5의 배수의 제곱:", squares_of_multiples_of_5)
```
