```python
age = 20
name = 'jennie'
str7 = '이름은 {:s} 나이는 {:d}'.format(name, age)
print(str7)
str8 = 'pi = {:10f}'.format(3.141592653589793)
print(str8)
str9 = 'pi = {:10.2f}'.format(3.141592653589793)
print(str9)
print('pi = %.2f' % 3.141592653589793)
```
